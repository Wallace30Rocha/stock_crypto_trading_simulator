## Apple

```python
#choose ticker from yahoo finance website
apple = yf.Ticker('AAPL')

#get the table including the oldest data until current time
apple_df = apple.history(period = 'max')

#insert a new column called Ticker in the first position with all rows set to "stock_ticker"
apple_df.insert(0, 'ticker', 'AAPL')

#transform the index 'Date' in a column
apple_df = apple_df.reset_index()

#eliminate the time from the 'Date' column
apple_df['Date'] = pd.to_datetime(apple_df['Date'])
apple_df['Date'] = apple_df['Date'].dt.date

#rename column Date
apple_df = apple_df.rename(columns = {'Date': 'date', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close' })

#create a new column with % change from last close to new open
apple_df['percent_change'] = (apple_df['close'] - apple_df['close'].shift(1)) / apple_df['close'].shift(1) * 100

#eliminate columns with no use
apple_df = apple_df.drop(['Volume', 'Dividends', 'Stock Splits'], axis = 1)

#convert NaN in % column to 0
apple_df['percent_change'] = apple_df['percent_change'].fillna(0)

#convert inf values to 0
apple_df['percent_change'] = apple_df['percent_change'].replace([np.inf, -np.inf], 0)

apple_df.head()
```
